```python
import requests
```

```python
offset_ = 0
limit_ = 1000
amount_ = 11 #всего 10522 объекта, для цикла вводим переменную, достигнем всех объектов на 11 итерации, если берем по 1000 объектов за раз
```

```python
objects_list = []
for i in range(0, amount_): #цикл идет, пока не переберем все 11 тыс объектов
    url = f'https://xn--80az8a.xn--d1aqf.xn--p1ai/%D1%81%D0%B5%D1%80%D0%B2%D0%B8%D1%81%D1%8B/api/kn/object?offset={offset_}&limit={limit_}&sortField=devId.devShortCleanNm&sortType=asc&objStatus=0'
    res = requests.get(url)
    objects_data = res.json()
    objects_list.extend(objects_data.get('data').get('list'))
    objects_data.get('data').get('list')[0]
    offset_ += amount_
    #print(objects_data) обрабатывается очень долго, поэтому закомментила
```

```python
objids = [x.get('objId') for x in objects_list]
print(objids)
len(objids)
```
